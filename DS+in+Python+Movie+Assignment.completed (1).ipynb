- ### Subtask 3.5: Find the best directors

    1. Group the dataframe using the `director_name` column.
    2. Find out the top 10 directors for whom the mean of `imdb_score` is the highest and store them in a new dataframe `top10director`. 

```python
# Write your code for extracting the top 10 directors here
import pandas as pd

# Assuming you have a DataFrame 'df' with columns 'director_name' and 'imdb_score'

# Step 1: Group the dataframe by 'director_name'
grouped_directors = df.groupby('director_name')

# Step 2: Calculate the mean IMDb score for each director
director_mean_scores = grouped_directors['imdb_score'].mean()

# Step 3: Sort the directors based on the mean IMDb score in descending order
sorted_directors = director_mean_scores.sort_values(ascending=False)

# Step 4: Extract the top 10 directors with the highest mean IMDb scores
top10director = sorted_directors.head(10).reset_index()

# Rename the columns for clarity
top10director.columns = ['director_name', 'mean_imdb_score']

# Display the top 10 directors
print(top10director)

```

Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.
    4. Find the mean of the `num_critic_for_reviews` and `num_user_for_review` and identify the actors which have the highest mean.

```python
# Write your code for creating three new dataframes 
Meryl_Streep = movies[movies['actor_1_name'] == 'Meryl Streep']
```

```python
Leo_Caprio = movies[movies['actor_1_name'] == 'Leonardo DiCaprio']

# Include all movies in which Leo_Caprio is the lead
```

```python
Brad_Pitt= movies[movies['actor_1_name'] == 'Brad Pitt'] # Include all movies in which Brad_Pitt is the lead
```

```python

# Write your code for combining the three dataframes here
# Combine the dataframes
Combined = pd.concat([Meryl_Streep, Leo_Caprio, Brad_Pitt], ignore_index=True)

# Check the first few rows of the combined dataframe
print("Combined Dataframe:\n", Combined.head())
```

```python
# Write your code for grouping the combined dataframe here
# Group by actor_1_name
grouped = Combined.groupby('actor_1_name')
```

```python
# Write the code for finding the mean of critic reviews and audience reviews here
# Calculate the mean of the relevant columns
mean_reviews = grouped[['num_critic_for_reviews', 'num_user_for_reviews']].mean()

# Display the mean reviews
print("Mean Reviews:\n", mean_reviews)
```

```python
# Identify the actor with the highest mean reviews
critic_favorite = mean_reviews['num_critic_for_reviews'].idxmax()
audience_favorite = mean_reviews['num_user_for_reviews'].idxmax()

# Display the results
print(f"The critic-favorite actor is: {critic_favorite}")
print(f"The audience-favorite actor is: {audience_favorite}")
```
